```python
import pandas as pd

#Cargar y solucionar el problema de la coma por punto, sin modificar nada
df = pd.read_csv("dataset_set_A_aguas_residuales.csv", sep=';', decimal=',')

# Convertir la fecha a datetime
df['fecha_registro'] = pd.to_datetime(df['fecha_registro'])

#Separar los datos para el area de operaciones
columnas_operaciones = [
    'fecha_registro', 'planta', 'caudal_entrada_m3_d',
    'DBO_entrada_mg_L', 'DBO_salida_mg_L',
    'energia_aeracion_kWh', 'lodos_generados_kg_d'
]
df_operaciones = df[columnas_operaciones]
df_operaciones.to_csv("reporte_area_operaciones.csv", index=False, sep=';', decimal=',')

#Separar los datos para el area de gestion ambiental
columnas_ambiental = [
    'fecha_registro', 'planta', 'DBO_salida_mg_L', 'cumplimiento_norma'
]
df_ambiental = df[columnas_ambiental]
df_ambiental.to_csv("reporte_ambiental.csv", index=False, sep=';', decimal=',')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import joblib

# 1. Cargar el dataset
df = pd.read_csv("dataset_set_A_aguas_residuales.csv", sep=';', decimal=',')

#Grafico de dispersion para ver la relacion del DBO con el cumplimiento
plt.figure()
sns.scatterplot(data=df, x="DBO_entrada_mg_L", y="DBO_salida_mg_L", hue="cumplimiento_norma", palette="viridis")
plt.title("Relación DBO Entrada vs DBO Salida según cumplimiento")
plt.show()


```
